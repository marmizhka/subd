Изначально представленная таблица нормализована до 1НФ, поскольку в ячейках таблицы нет повторяющихся значений. Заметим, что в таблице некоторые данные (brand, product_line и др) зависят от product_id, а не от transaction_id, из-за чего многие строчки повторяются, меняется только transaction_id. Чтобы нормализовать таблицу до 2НФ, необходимо разделить ее на 2 таблицы: transactions и products. Ниже реализовано разделение на таблицы transactions и products. Теперь вся информация о продуктах содержится в табл products и не дублируется в transactions.

Значит, для нормализации ттаблицы до 3НФ, нужно вынести customers в отдельную табл и сделать ключом customer_id. Тогда транзитивность будет устранена.
